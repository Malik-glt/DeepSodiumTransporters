```python
import numpy as np
import os

def concatenate_and_save_data(train_sat_data_paths, test_slc_data_paths, train_sat_label_paths, test_slc_label_paths, output_folder):
    # Load and concatenate training SAT data
    x_train_sat_list = [np.load(path) for path in train_sat_data_paths]
    x_train_sat = np.concatenate(x_train_sat_list, axis=0)

    # Load and concatenate testing SLC data
    x_test_slc_list = [np.load(path) for path in test_slc_data_paths]
    x_test_slc = np.concatenate(x_test_slc_list, axis=0)

    # Save concatenated SAT data
    np.save(os.path.join(output_folder, 'SATS_data.npy'), x_train_sat)

    # Save concatenated SLC data
    np.save(os.path.join(output_folder, 'SLC_data.npy'), x_test_slc)

    # Load and concatenate training SAT labels
    y_train_sat_list = [np.load(path) for path in train_sat_label_paths]
    y_train_sat = np.concatenate(y_train_sat_list, axis=0)

    # Load and concatenate testing SLC labels
    y_test_slc_list = [np.load(path) for path in test_slc_label_paths]
    y_test_slc = np.concatenate(y_test_slc_list, axis=0)

    # Save concatenated SAT labels
    np.save(os.path.join(output_folder, 'SATS_labels.npy'), y_train_sat)

    # Save concatenated SLC labels
    np.save(os.path.join(output_folder, 'SLC_labels.npy'), y_test_slc)

if __name__ == "__main__":
    train_sat_data_folder = 'D:/Malik/Functional_Identification/get_feature/pssm_d20_L1218/SATs'
    test_slc_data_folder = 'D:/Malik/Functional_Identification/get_feature/pssm_d20_L1218/SLC'
    train_sat_label_folder = 'D:/Malik/Functional_Identification/get_feature/pssm_d20_L1218/SATs_labels'
    test_slc_label_folder = 'D:/Malik/Functional_Identification/get_feature/pssm_d20_L1218/SLC_labels'

    train_sat_data_paths = [os.path.join(train_sat_data_folder, file) for file in os.listdir(train_sat_data_folder) if file.endswith('.npy')]
    test_slc_data_paths = [os.path.join(test_slc_data_folder, file) for file in os.listdir(test_slc_data_folder) if file.endswith('.npy')]
    train_sat_label_paths = [os.path.join(train_sat_label_folder, file) for file in os.listdir(train_sat_label_folder) if file.endswith('.npy')]
    test_slc_label_paths = [os.path.join(test_slc_label_folder, file) for file in os.listdir(test_slc_label_folder) if file.endswith('.npy')]

    output_folder = 'D:/Malik/SATs/Amino acid vs others/Prottrans'
    
    concatenate_and_save_data(train_sat_data_paths, test_slc_data_paths, train_sat_label_paths, test_slc_label_paths, output_folder)
```

# Sodium class

```python
import numpy as np
import os

def concatenate_and_save_data(data_folder, output_folder):
    train_data_paths = []
    test_data_paths = []
    train_labels = []
    test_labels = []

    for class_folder in os.listdir(data_folder):
        class_folder_path = os.path.join(data_folder, class_folder)
        if os.path.isdir(class_folder_path):
            train_folder_path = os.path.join(class_folder_path, 'Train')
            test_folder_path = os.path.join(class_folder_path, 'Test')

            # Collect train data paths
            train_data_paths.extend([os.path.join(train_folder_path, file) for file in os.listdir(train_folder_path) if file.endswith('.npy')])

            # Collect test data paths
            test_data_paths.extend([os.path.join(test_folder_path, file) for file in os.listdir(test_folder_path) if file.endswith('.npy')])

            # Generate and collect train labels
            train_labels.extend([1 if 'Sodium' in class_folder else 0] * len(os.listdir(train_folder_path)))

            # Generate and collect test labels
            test_labels.extend([1 if 'Sodium' in class_folder else 0] * len(os.listdir(test_folder_path)))

    # Load and concatenate training data
    x_train_list = [np.load(path) for path in train_data_paths]
    x_train = np.concatenate(x_train_list, axis=0)

    # Save concatenated training data
    np.save(os.path.join(output_folder, 'Train_data.npy'), x_train)

    # Save concatenated training labels
    np.save(os.path.join(output_folder, 'Train_labels.npy'), np.array(train_labels))

    # Load and concatenate testing data
    x_test_list = [np.load(path) for path in test_data_paths]
    x_test = np.concatenate(x_test_list, axis=0)

    # Save concatenated testing data
    np.save(os.path.join(output_folder, 'Test_data.npy'), x_test)

    # Save concatenated testing labels
    np.save(os.path.join(output_folder, 'Test_labels.npy'), np.array(test_labels))

if __name__ == "__main__":
    data_folder = 'D:/Malik/SATs/Sodium vs others/get_feature/binary_d20_L1200'
    output_folder = 'D:/Malik/SATs/Sodium vs others/Binary'

    concatenate_and_save_data(data_folder, output_folder)
```

# Concatinate SLC

```python
import numpy as np
import os

def concatenate_and_save_independent_data(data_folder, output_folder):
    independent_data_paths = []
    independent_labels = []

    for class_folder in os.listdir(data_folder):
        class_folder_path = os.path.join(data_folder, class_folder)
        if os.path.isdir(class_folder_path):
            # Collect independent data paths
            independent_data_paths.extend([os.path.join(class_folder_path, file) for file in os.listdir(class_folder_path) if file.endswith('.npy')])

            # Generate and collect independent labels
            independent_labels.extend([1 if 'Sodium' in class_folder else 0] * len(os.listdir(class_folder_path)))

    # Load and concatenate independent data
    independent_data_list = [np.load(path) for path in independent_data_paths]
    independent_data = np.concatenate(independent_data_list, axis=0)

    # Save concatenated independent data
    np.save(os.path.join(output_folder, 'Independent_data.npy'), independent_data)

    # Save concatenated independent labels
    np.save(os.path.join(output_folder, 'Independent_labels.npy'), np.array(independent_labels))

if __name__ == "__main__":
    data_folder = 'D:/Malik/SATs/Sodium vs others/get_feature/pt_d1024_L1200/SLC'
    output_folder = 'D:/Malik/SATs/Sodium vs others/SLC'

    concatenate_and_save_independent_data(data_folder, output_folder)
```
